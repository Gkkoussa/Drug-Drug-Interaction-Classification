```python
import json
import pandas as pd
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from adjustText import adjust_text
import umap
import plotly.express as px
from sklearn.decomposition import PCA
```

```python
with open('/Users/onature/Downloads/interactions.json') as f:
    drug_interactions = json.load(f)
```

```python
drugs = set(i for i in drug_interactions.keys())
for j in drug_interactions.values():
    drugs.update(j)
```

```python

adjacency_matrix = pd.DataFrame(0, index= drugs, columns= drugs)#an empty data frame with drugs as index and columns
```

```python
for drug, interacting_drugs in drug_interactions.items():
    for i in interacting_drugs:
        adjacency_matrix.at[drug, i] = 1
        adjacency_matrix.at[i, drug] = 1
```

```python
umapreducer = umap.UMAP(random_state=42, n_jobs=-1)
fittedumap = umapreducer.fit_transform(adjacency_matrix)
```

```python
umap_df = pd.DataFrame(fittedumap, columns=['UMAP Component 1', 'UMAP Component 2'])
umap_df['Drug'] = adjacency_matrix.columns


fig = px.scatter(umap_df, x='UMAP Component 1', y='UMAP Component 2', hover_name='Drug', title='UMAP Visualization of Drug Interactions')

fig.show()
```

UMAP appears to do a much better joke clustering the different drugs. This makes sense given we found the categorical features to be most important

```python
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
fittedtsne = tsne.fit_transform(adjacency_matrix)
```

```python
tsne_df = pd.DataFrame(fittedtsne, columns=['t-SNE Component 1', 't-SNE Component 2'])
tsne_df['Drug'] = adjacency_matrix.columns


fig = px.scatter(tsne_df, x='t-SNE Component 1', y='t-SNE Component 2', hover_name='Drug', title='t=SNE Visualization of Drug Interactions')

fig.show()
```

```python
pca = PCA(n_components=2)
fittedpca = pca.fit_transform(adjacency_matrix)
```

```python
pca_df = pd.DataFrame(fittedpca, columns=['PCA Component 1', 'PCA Component 2'])
pca_df['Drug'] = adjacency_matrix.columns


fig = px.scatter(pca_df, x='PCA Component 1', y='PCA Component 2', hover_name='Drug', title='PCA Visualization of Drug Interactions')

fig.show()
```

Interestingly there appear to be 3 or 4 distinct streaks of drugs. It may be worth looking into what each of these 3/4 classes of drugs represent.
